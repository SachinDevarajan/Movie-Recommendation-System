```python
import joblib
import faiss
from sentence_transformers import SentenceTransformer
```

```python
df = joblib.load('movies.pkl')
index = faiss.read_index("movies.index")
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
def answer(question,k=3):
    que_embed = model.encode([question])
    distances, indices = index.search(que_embed,k)
    
    results = []
    for i in indices[0]:
        title = df.iloc[i]["title"]
        overview = df.iloc[i]["overview"]
        results.append({
            "title": title,
            "overview": overview
            })

    return results
```

```python
question=input("Enter Movie Type : ")
question
```

```python
movies = answer(question)
for i in movies:
    print(i)
```
